```python
def primary_taxonomy(df):
    for col in range(1, 16):
        if df['Healthcare Provider Primary Taxonomy Switch_' + str(col)] == 'Y':
            return df['Healthcare Provider Taxonomy Code_' + str(col)]
```

```python
nppes_taxonomy['primary_taxonomy'] = nppes_taxonomy.apply(lambda x: primary_taxonomy(x), axis = 1)
```
