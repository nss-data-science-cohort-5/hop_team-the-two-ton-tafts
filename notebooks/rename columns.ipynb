```python
import pandas as pd
```

```python
nppes = pd.read_csv('../data/nppes/nppes_trimmed.csv')
```

```python
nppes.info()
```

```python
cols = {
    'NPI': 'npi',
    'Entity Type Code': 'entity_code',
    'Provider Organization Name (Legal Business Name)': 'org_name',
    'Provider Last Name (Legal Name)': 'last_name',
    'Provider First Name': 'first_name',
    'Provider Middle Name': 'middle',
    'Provider Name Prefix Text': 'prefix',
    'Provider Name Suffix Text': 'suffix',
    'Provider Credential Text': 'credential',
    'Provider First Line Business Practice Location Address': 'address1',
    'Provider Second Line Business Practice Location Address': 'address2',
    'Provider Business Practice Location Address City Name': 'city',
    'Provider Business Practice Location Address State Name': 'state',
    'Provider Business Practice Location Address Postal Code':'postal_code'
}
```

```python
nppes = nppes.rename(columns = cols)
```

```python
nppes
```

```python
taxonomy = pd.read_csv('../data/nppes/nppes_taxonomy.csv')
```

```python
taxonomy
```

```python
nppes = nppes.merge(taxonomy,
           left_on = 'npi',
           right_on = 'NPI')
```

```python
nppes = nppes.drop('NPI',
          axis = 1)
```

```python
nppes = nppes.dropna(subset = ['primary_taxonomy'])
```

```python
nppes.to_csv('../data/nppes/nppes_with_taxonomy.csv',
            index = False)
```
