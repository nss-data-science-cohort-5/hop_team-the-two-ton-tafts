```python
import pandas as pd
import sqlite3
from tqdm.notebook import tqdm
import numpy as np
```

```python
vandy_recs = pd.read_csv('data/prov_org_vandy_perc.csv')
```

```python
vandy_recs
```

```python
vandy_recs['total'] = vandy_recs['non_vandy_transactions'] + vandy_recs['vandy_transactions']
vandy_recs=vandy_recs.drop(columns='Unnamed: 0')
vandy_recs
```

```python
hop_proj=sqlite3.connect('data/hop_proj.sqlite')

query = """
SELECT DISTINCT class_spec
FROM taxonomy
JOIN nppes
ON code = primary_taxonomy
WHERE org_name LIKE '%vanderbilt%'
    
"""

vandy_specs = pd.read_sql(query, hop_proj)

vandy_specs
```

```python
vandy_services = vandy_specs['class_spec'].tolist()

vandy = vandy_recs.columns
vandy_df = pd.DataFrame(columns = vandy)
for i in vandy_services:
    result = vandy_recs[vandy_recs['org_spec'] == i]
    vandy_df = vandy_df.append(result)
```

```python
vandy_df.sort_values(by = 'org_spec')
```

```python
vandy_df.pivot_table(values='total', index=['prov_zip', 'org_spec'], aggfunc=np.sum, fill_value=0).reset_index()
```
